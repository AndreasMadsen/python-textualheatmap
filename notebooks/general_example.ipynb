```python
try:
    %tensorflow_version 2.x
except:
    pass

import json
import urllib.request as request
from textualheatmap import TextualHeatmap
```

```python
try:
    with open('./heatmap.json') as fp:
        data = json.load(fp)
except IOError:
    with request.urlopen('https://raw.githubusercontent.com/AndreasMadsen/python-textualheatmap/master/notebooks/heatmap.json') as fp:
        data = json.load(fp)        
```

```python
heatmap = TextualHeatmap(facet_titles = ['GRU', 'LSTM'], show_meta=True)
heatmap.set_data(data)
heatmap.highlight(159)
```

```python
heatmap = TextualHeatmap(
    facet_titles = ['GRU', 'LSTM'],
    rotate_facet_titles=True
)
heatmap.set_data(data)
heatmap.highlight(159)
```
